```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = {
    'Produk': ['A', 'B', 'C', 'A', 'B', 'C'],
    'Penjualan': [1500, 2300, 1800, 1700, 2100, 1900],
    'Bulan': ['Jan', 'Jan', 'Jan', 'Feb', 'Feb', 'Feb']
}
df = pd.DataFrame(data)
sns.set_style("whitegrid")
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x='Produk', y='Penjualan', data=df, estimator=sum)
plt.title('Penjualan per Produk')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.lineplot(x='Bulan', y='Penjualan', data=df, hue='Produk', marker='o')
plt.title('Tren Penjualan Bulanan')
plt.show()
```
