```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('data_pejualan_day1.csv')
df.columns = df.columns.str.strip().str.lower()

# Grouping total penjualan per produk
df_produk = df.groupby('produk')['total'].sum().sort_values(ascending=False)

df_produk.plot(kind='bar', figsize=(8,5))
plt.title('Total Penjualan per Produk')
plt.xlabel('Produk')
plt.ylabel('Total Penjualan (Rp)')
plt.grid(axis='y')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
